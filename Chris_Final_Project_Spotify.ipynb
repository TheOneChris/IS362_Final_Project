```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
top200music = pd.read_csv('https://raw.githubusercontent.com/TheOneChris/IS362_Final_Project/main/top100_spotify_2022.csv', index_col=False)
spotify = pd.DataFrame(top200music)
spotify
```

```python
#We then get / display the details behind the data (mean, count, standard deviation, min/maxn, etc..)
attributes = spotify.describe()
attributes
```

```python
#Now lets start graphing, looking at some histograms of all variables
top200music.hist(figsize= [15, 15])
```

```python
#Transform the data to fit what we want to analyze
spotify.dropna()
spotify.drop(['added', 'dnce', 'live', 'val', 'acous', 'spch', 'pop', 'explicit'], axis=1)
```

```python
#Which artists had the most top 100 hits in 2022 so far?
top200music['Frequency']=1
top200music_artist=top200music.groupby('Artist')['Frequency'].sum().reset_index().sort_values(by='Frequency',ascending=False)
top200music_artists_ten = top200music_artist.head(10)
top200music_artists_ten
```

```python
art_bar = sns.barplot(x="Artist", y="Frequency", color="Blue", data=top200music_artists_ten)
art_bar.set_xticklabels(art_bar.get_xticklabels(), rotation=30)
```

Post Malone leads with 6 songs in the 2022 top 100! Ed Sheeran came in a close 2nd with 5 songs. 

Now let's start analyzing the mean value from the song data! 

Starting with the BPM of the songs on the top 100 list

```python
print('Mean value for BPM:', spotify['BPM'].mean())
sns.displot(spotify['BPM'])
plt.xlabel('BPM',size=20)
plt.show()
```

Next, let's analyze the mean value for song Energyof songs on the top 100 list

```python
print('Mean value for Energy:', spotify['Energy'].mean())
sns.displot(spotify['Energy'])
plt.xlabel('Energy',size=20)
plt.show()
```

Next, let's analyze the mean value for song Decibels of songs on the top 100 list

```python
print('Mean value for Decibel:', spotify['Decibel'].mean())
sns.displot(spotify['Decibel'])
plt.xlabel('Decibel',size=20)
plt.show()
```

Next, let's analyze the mean value for song Duration of songs on the top 100 list

```python
print('Mean value for song Duration:', spotify['Duration'].mean())
sns.displot(spotify['Duration'])
plt.xlabel('Duration',size=20)
plt.show()
```

Lastly, we will combine these statistics into a single graph to compare how certain features may make certain songs more popular than others

```python
data=spotify.drop(['Year', 'dnce', 'live', 'val', 'acous', 'spch', 'pop', 'explicit', 'Frequency'], axis=1)
data.mean().plot.bar()
plt.title('Mean Values of Audio Features')
plt.show()
```

From the graph, we can summarize that most of the top songs streamed on spotify for year 2022 so far were at least just over 200 seconds in length (~3 minutes long) and have a fairly high BPM (faster pace) and Energy factors. This would lead me to believe many "Dance" or "Rap" music songs being the the most popular for streaming on spotify.
